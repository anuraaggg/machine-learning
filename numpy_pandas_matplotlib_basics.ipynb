Matplotlib Basics

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.plot(x, y)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Simple Line Plot")
plt.show()
```

```python
plt.scatter(x, y)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Scatter Plot")
plt.show()
```

```python
students = ["Devi","Vijay","Kumar","Meena"]
marks = [85,92,78,88]
plt.bar(students, marks)
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Students vs Marks")
plt.show()
```

```python
data = np.random.randn(1000)
plt.hist(data, bins=30)
plt.title("Histogram Example")
plt.show()
```

```python
sizes = [40,30,20,10]
labels = ["Python","Java","C","Others"]
plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.title("Programming Language Usage")
plt.show()
```

```python
plt.plot(x, y, color='red', marker='o', linestyle='--')
```

```python
plt.grid(True)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("Line")
plt.subplot(1,2,2) 
plt.bar(students,marks)
plt.title("Bar")
plt.show()
```
